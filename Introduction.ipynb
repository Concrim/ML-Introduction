 # 1. Machine Learning

### What is Machine Learning?
It is to make a computer perform a task or improve without any explicit programming.

### Why Machine Learning?
<ol>
    <li>It is preferable over traditional programming where the problem can change over and over. We may only have to use a single Algorithm.</li>
    <li>Data Mining. We can gather new insights and patterns through ML.</li>
    <li>It is impractical to use traditional programming for a wide number of problems, which can be countered by ML.</li>
</ol>

# 2. AI vs. ML vs. DL
In contrast with AI, ML does not have to be explicitly programming, nor do we require a prior set of rules. We are only in need for abundant data and the algorithms will derive rules automatically and find distinctions.

#### What is Deep Learning?
DL is a branch of Machine Learning, inspired by the human brain. It is based around the idea of perceptrons. It shines in areas where Machine Learning might lack.

In Machine Learning, we need to define features in order to get the required results. However, DL is not in need of such definitions and can detect these patterns by itself. In certain cases where we don't know what features can be used, DL proves highly effective.

In DL, we can provide raw data and the model derives the features automatically. The more the data, the better the DL model performs, whereas ML plateaus.

#### <u>But:</u>
<ul>
    <li>DL is effective for large data.</li>
    <li>DL is computationally expensive.</li>
</ul>

# 3. Types of Machine Learning

### 1. Supervised
If we are working with labeled data, it's known as Supervised learning.
<ul><li><b>Regression</b></li>Output is numerical.<li><b>Classification</b></li>Output is categorical.</ul>

### 2. Unsupervised
If we are working with unlabeled data, it's known as Unsupervised learning.
<ul><li><b>Clustering</b></li>We categorize similar outputs together.<li><b>Dimensionality Reduction</b></li>We remove extra columns/dimensions from our input to reduce cost and improve visualization e.g. PCA<li><b>Anomaly Detection</b></li> We detect and analyze anomalies within our data.<li><b>Association</b></li>We detect hidden patterns and relations within our data.</ul>

### 3. Semi-supervised
This type of learning is partially supervised and partially unsupervised, e.g. Google Photos people classification.

### 4. Reinforcement
This type of learning uses a reward system to find the best course of action. The <i>agent</i> is provided with a rulebook and updates its <i>policy</i> in such a way that it completes its task effectively.

# 4. Batch Learning 

In Batch Learning, you take your data and train your model on it while offline and then you upload it on the server.
### Problems:
<ul>
    <li>This type of learning is static, so it cannot evolve with developments.</li><li>With new data, we have to retrain it every time.</li><li>If amount of data is too large, we might not be able to handle it.</li><li>Hardware Limitation.</li><li>Availability Issues: While retraining, we won't have access to the model.</li>
</ul>


# 5. Online Learning
"The more you use it, the more it'll improve."<br>
The training is done incrementally through mini-batches while the model is <i>on the server</i>.<br>
The model is training and predicting at the same time. <br>
E.g. YouTube, Typing Recommendations.



### Why to use?
<ul>
    <li>Where there is a concept drift e.g. Stock Exchange</li>
    <li>Cost-Effective</li>
    <li>Faster Solution</li>
</ul>

### How to implement?
<ul>
    <li>Partial fit.</li><li>Libraries such as River, Vowpal Rabbit</li>
</ul>

### Learning Rate
This is how often you update the model.<br>
Must select optimum rate, or model will either be unupdated or forget prior information.

### Out of Core Learning
If the dataset is too huge, we split it into smaller chunks, train the model separately, and then upload it.

### Problems
<ul>
    <li>Tricky to Use
    </li>
    <li>Risky to Use</li>
</ul>

# Batch Learning vs Online Learning
<ol>
    <li><b>Complexity</b>: Offline is less complex as model is constant, while Online has dynamic complexity.</li>
    <li><b>Computational Power</b>: Offline has fewer computations and one time training. Online has continous data ingestion, so high compuatation.</li>
    <li><b>Production</b>: Offline is easier to implement. Online is difficult to implement and manage.</li>
    <li><b>Applications</b>: Offline is used in Image Classification or Machine Learning where data patterns remain constant. Online is used in Finance, Economics, where new patterns constantly emerge.</li>
    <li><b>Tools</b>: Offline uses Scikit, Tensorflow, Pytorch, Keras. Online used Active Research, MOA, SAMOA, scikit-multiflow.</li>
</ol>

# 6. Instance-Based Learning vs Model-Based Learning
### Instance-Based
This kind of learning basically 'crams' the data.
<br>We use the concept of similarity with other data-points, e.g. KNN.
### Model-Based
This kind of Learning works with patterns within the data.
<br>We use internal information within the data, such as a boundary or a mathematical function. After this, we are not it need of data-points, and can use the <i>decision function</i> to determine results, e.g. Linear Regression.
<br><br>
Most of the models we encounter are model-based.

### Differences
<ol>
    <li><b>Preparation:</b> Data has to be prepared for training in both.</li>
    <li><b>Usage: </b>In Model Based, we train model.<br> In Instance-Based, we detect patterns.</li>
    <li><b>Storage: </b>MB has to be stored.<br> IB has no model to store.</li>
    <li><b>Generalization: </b>MB has to be generalized before socring instance.<br> IB is generalized after each scoring instance.</li>
    <li><b>Prediction: </b>MB predicts using model.<br> IB predicts directly using training data.</li>
    <li><b>Discarding: </b>MB allows discarding training data. <br>IB requries the training data, so it can't be thrown away. </li>
    <li><b>Memory: </b>MB requires lesser memory.<br> IB requires large amount of memory.</li>
</ol>

# 7. Problems of Machine Learning
<ol>
    <li><b>Data Collection</b>: It is hard to obtain data. We need APIs or Webscraping, but they themselves have issues.</li>
    <li><b>Insufficient Data</b>: We don't always have abundant enough amounts of data required</li>
    <li><b>Non-Representative Data: </b>If the data is incomplete or biased, we will have a skewed model, e.g. Sampling Noise, Sampling Bias.</li>
    <li><b>Poor Qaulity Data: </b>We spend 60% of the time preprocessing the data, e.g. Outliers, Null values, etc.</li>
    <li><b>Irrelevant Features</b>: Garbage in, Garbage Out. We need to remove irrelevant features for the model to perform well.</li>
    <li><b>Overfitting</b>: A model that works very well on the training data but fails on new data has overfit.</li>
    <li><b>Underfitting</b>: This is the opposite of Overfitting. The model is oversimplified and does not perform well.</li>
    <li><b>Software Integration</b>: Making a model is not enough. We need to make it work with machines and computers.</li>
    <li><b>Offline Learning: </b>Flaws of Offline Learning are discussed in section 4. Deployment is very costly.</li>
    <li><b>Cost Involved</b>: If you are working large-scale, there will be a lot of expenses.</li>
</ol>

<b>It is recommended to deploy your models/projects in order to maximize learning.</b>

# 9. Machine Learning Life Cycle
This focuses on the <b>how</b> of Machine Learning.
### MLDLC
We need to focus on making end-to-end products.

<ol>
    <li><b>Frame the Problem: </b>Before starting on your product, you need to go over the details of your product such as client, intricacies of project, timeline, etc.</li>
    <li><b>Gathering Data: </b>ML is incomplete without Data. We need to collect data from different sources, e.g. CSV, API, webscraping, data warehouse, etc.</li>
    <li><b>Preprocessing: </b>Exported data is bound to be unclean and problematic. We need to prepare this data thoroughly so it can be used, e.g. Duplicates, Missing Values, Outliers, Scaling, etc.</li>
    <li><b>EDA: </b>We anaylyse our data and visualize it to gather insights, e.g. Outlier Detection, Imbalance, Patterns, etc.</li>
    <li><b>Feature Engineering: </b>We need to create new columns and select appropriate ones to help us perform better on the data, e.g. Combining features to make new feature.</li>
    <li><b>Model Training, Evaluation, and Selection: </b>In this stage, we try out different models and go through them. Ultimately, we select the one which is most efficient.</li>
    <li><b>Deployment: </b>We need to convert out model into a software to make it accessible to the clients/customers. </li>
    <li><b>Testing: </b>Before complete deployment, we do selective (Beta) testing to gather reviews and feedback for our product. Accordinly, we can make changes to our product.</li>
    <li><b>Optimizing: </b>This is the process of maintaining the quality of your product. We take steps to ensure the product does not <i>rot</i>.</li>
</ol>
